Create the paraphrase dataset to finetune the GPT-2 Model for paraphrasing

The dataset will combine different datasets from different sources
1. PAWS Dataset. [Link Here](https://huggingface.co/datasets/paws).
2. Quora Dataset. [Link Here](https://huggingface.co/datasets/quora).
3. TaPaCo (en) Dataset. [Link Here](https://zenodo.org/record/3707949#.Y3tNYC2B35b).

```python
from datasets import load_dataset
import pandas as pd
```

```python
paws_data = load_dataset("paws", "labeled_final")
```

```python
paws_df = pd.DataFrame(paws_data["train"])
```

```python
paws_df_filtered = paws_df[paws_df["label"]==1]
paws_df_filtered
```

```python
quora_data = load_dataset("quora")
```

```python
quora_df = pd.DataFrame(quora_data["train"])
```

```python
quora_df_filtered = quora_df[quora_df["is_duplicate"]==True]
quora_df_filtered
```

```python
def quora_to_sentence_pair(d):
       text = d["text"]
       return text[0], text[1]
```

```python
quora_df_filtered["sentence1"], quora_df_filtered["sentence2"] = zip(*quora_df_filtered['questions'].map(quora_to_sentence_pair))
```

```python
quora_df_filtered
```

```python
tapaco_dataset = load_dataset("tapaco", "en")
```

```python
tapaco_df = pd.DataFrame(tapaco_dataset["train"])
```

```python
tapaco_df
```

```python
# create sentence pairs within each paraphrase set (paraphrase sets with the same id)
prev_id = None
paraphrase_stack = []
paraphrases = []
for index, row in tapaco_df.iterrows():
       paraphrase_id, text = row["paraphrase_set_id"], row["paraphrase"]
       if prev_id==paraphrase_id:
              paraphrase_stack.append(text)
              if len(paraphrase_stack)>1:
                     paraphrases.append(tuple(paraphrase_stack))
              paraphrase_stack = []
              prev_id=None
       else:
              prev_id = paraphrase_id
              paraphrase_stack = [text]
```

```python
tapaco_paraphrases = pd.DataFrame(paraphrases, columns=["sentence1", "sentence2"])
```

```python
tapaco_paraphrases
```

```python
tapaco_2_column = tapaco_paraphrases[["sentence1", "sentence2"]].reset_index()
quora_2_column = quora_df_filtered[["sentence1", "sentence2"]].reset_index()
paws_2_column = paws_df_filtered[["sentence1", "sentence2"]].reset_index()
```

```python
total = pd.concat([tapaco_2_column, quora_2_column, paws_2_column], axis=0).reset_index()
```

```python
total.drop(["level_0", "index"], axis=1, inplace=True)
```

```python
total
```

```python
# research ways to prompt gpt to paraphrase
total["combined"] = '<s>'+total.sentence1+'</s>'+">>>>"+"<p>"+total.sentence2+"</p>"
```

```python
total["combined"] = total.combined.to_csv("combined.txt", sep="\n", index=False)
```
